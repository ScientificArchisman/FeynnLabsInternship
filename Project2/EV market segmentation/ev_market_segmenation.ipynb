## Using Unsupervised KNN Clustering

Here's our plan:

1. **Data Preparation**: We'll prepare our dataset by selecting relevant features and scaling them since clustering algorithms are sensitive to the scale of data.
2. **Determine Optimal Number of Clusters**: Using the Elbow Method and Silhouette Score.
3. **K-means Clustering**: Apply the algorithm using the optimal number of clusters.
4. **Analyze Clusters**: Examine the characteristics of each cluster to determine its defining features.

Let's start with **Data Preparation**. We'll consider features like `PriceEuro`, `Range`, and `Efficiency` for clustering.

```python
from sklearn.preprocessing import StandardScaler

# Selecting relevant features for clustering
features = data[['PriceEuro', 'Range', 'Efficiency']]

# Scaling the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

scaled_features[:5]  # Displaying the first 5 rows of scaled features
```


The data has been successfully scaled, with each feature now having a mean of 0 and a standard deviation of 1.

Next, we'll determine the optimal number of clusters using the **Elbow Method** and **Silhouette Score**.

1. The **Elbow Method** involves running the K-means clustering on the dataset for a range of values of \( k \) and then plotting the Sum of Squared Errors (SSE) for each value of \( k \). The "elbow" of the curve represents an optimal value for \( k \) (a balance between precision and computational cost).
   
2. The **Silhouette Score** measures how close each point in one cluster is to the points in the neighboring clusters. Its values range from -1 to 1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.

Let's proceed with these methods to determine the optimal \( k \).

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Elbow Method
sse = []
for k in range(1, 11):  # We'll check for up to 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=42, n_init="auto")
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)

# Silhouette Score
sil_scores = []
for k in range(2, 11):  # Silhouette score is not defined for k=1
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(scaled_features)
    cluster_labels = kmeans.labels_
    sil_scores.append(silhouette_score(scaled_features, cluster_labels))

# Plotting the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Elbow Method
ax1.plot(range(1, 11), sse, marker='o')
ax1.set_title('Elbow Method')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('SSE')

# Silhouette Score
ax2.plot(range(2, 11), sil_scores, marker='o')
ax2.set_title('Silhouette Score')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Silhouette Score')

ax1.locator_params(axis = "both", nbins = 15)
ax2.locator_params(axis = 'both', nbins = 15)

plt.tight_layout()
plt.show()
```


The plots provide insights into the optimal number of clusters:

1. **Elbow Method**:
   - The plot shows a noticeable "elbow" around \( k = 4 \), suggesting that 4 might be a good choice for the number of clusters.
   
2. **Silhouette Score**:
   - The silhouette score seems to peak at \( k = 4 \) as well, further confirming that 3 clusters might be a good choice.

Given these results, we'll proceed with \( k = 4 \) clusters for our K-means clustering.

```python
# Applying K-means clustering with 4 clusters
kmeans_4 = KMeans(n_clusters=4, random_state=42, n_init="auto")
clusters_4 = kmeans_4.fit_predict(scaled_features)

# Adding the cluster labels to our original dataset
data['Cluster_4'] = clusters_4

# Analyzing the characteristics of each cluster with k=4
cluster_analysis_4 = data.groupby('Cluster_4').agg({
    'PriceEuro': ['mean', 'std'],
    'Range': ['mean', 'std'],
    'Efficiency': ['mean', 'std'],
    'Model': 'count'
}).reset_index()

cluster_analysis_4.columns = ['Cluster', 'Avg Price', 'Std Price', 'Avg Range', 'Std Range', 'Avg Efficiency', 'Std Efficiency', 'Number of Models']

cluster_analysis_4
```

Next, we'll follow these steps:

1. **Dimensionality Reduction**: Use Principal Component Analysis (PCA) to reduce the dimensionality of our data so we can visualize the clusters in a 2D space.
2. **Hierarchical Clustering**: Create a dendrogram to visualize and understand the hierarchical structure of the clusters.
3. **Cluster Profiling**: Use mosaic plots to profile the clusters based on different attributes.

Let's start with **Dimensionality Reduction using PCA** and visualize the clusters in a 2D space.

```python
from sklearn.decomposition import PCA

# Applying PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)

# Converting the principal components into a DataFrame for easier plotting
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Adding the cluster labels to the PCA DataFrame
pca_df['Cluster'] = clusters_4

# Plotting the 2D visualization of the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', palette='viridis', data=pca_df, s=60)
plt.title('2D PCA of EV Data with K-means Clustering (k=4)')
plt.show()
```


The 2D visualization using PCA shows the distribution of the four clusters. Each point represents a vehicle, and the colors represent the clusters to which they belong. The clusters appear to be reasonably well-separated, indicating that the K-means clustering with \( k = 4 \) has done a good job in segmenting the data.

Next, let's move on to **Hierarchical Clustering**. We'll create a dendrogram to understand the hierarchical structure of the clusters. This will help in visualizing the formation of clusters at various levels of granularity.

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Using the 'ward' method for hierarchical clustering
linked = linkage(scaled_features, method='ward')

# Setting up the figure size
plt.figure(figsize=(16, 8))

# Plotting the dendrogram
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance (Ward)')
plt.show()
```


The dendrogram provides a visual representation of the hierarchical clustering. Each line represents a merger of two clusters, and the height of the line represents the distance (or dissimilarity) at which the merger occurred.

The dendrogram can be interpreted as follows:
- Starting from the bottom, each leaf (or horizontal line) represents an individual data point (vehicle in our case).
- As we move up, similar data points (or clusters) merge, represented by the vertical lines.
- The height of the vertical lines indicates the dissimilarity at which the clusters merged. Higher vertical lines indicate greater dissimilarity.

By cutting the dendrogram horizontally at different heights, you can obtain different numbers of clusters. For instance, cutting it at the highest vertical line would give you two clusters, while cutting it lower could yield more clusters.

```python
# Grouping by the clusters to calculate the average values for the bubble plot
cluster_avg = data.groupby('Cluster_4').agg({
    'PriceEuro': 'mean',
    'Range': 'mean',
    'Efficiency': 'mean',
    'Model': 'count'
}).reset_index()

# Setting up the plot
plt.figure(figsize=(14, 10))

# Creating the bubble plot
scatter = plt.scatter(cluster_avg['PriceEuro'], cluster_avg['Range'], 
                      s=cluster_avg['Model']*100,  # Multiplying by 100 for better visualization
                      c=cluster_avg['Efficiency'], cmap='YlGnBu', edgecolors='black', linewidths=1.5, alpha=0.7)

# Annotations for cluster numbers
for i, txt in enumerate(cluster_avg['Cluster_4']):
    plt.annotate(txt, (cluster_avg['PriceEuro'].iloc[i], cluster_avg['Range'].iloc[i]), ha='center', va='center', color='red', 
                 fontsize=12, fontweight = "bold")

plt.xlim(20000)
plt.ylim(0)

plt.title('Bubble Plot of EV Clusters')
plt.xlabel('Average Price (Euros)')
plt.ylabel('Average Range (km)')
plt.colorbar(scatter, label='Average Efficiency (Wh/km)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

The bubble plot provides a comprehensive visualization of the clusters based on several attributes:

- **X-axis (Average Price)**: Indicates the average price of vehicles in each cluster.
- **Y-axis (Average Range)**: Represents the average range of vehicles in each cluster.
- **Size of the Bubble**: Denotes the number of models in each cluster, with larger bubbles indicating a higher number of models and thus potentially a more popular or dense segment.
- **Color**: Signifies the average efficiency of vehicles in each cluster, with darker colors indicating better efficiency.
